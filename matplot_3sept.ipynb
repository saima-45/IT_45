```python
import numpy as np
categories = ['1', '2', '3', '4']
sales_2024 = [100, 150, 120, 180]
sales_2025 = [130, 140, 160, 170]
x = np.arange(len(categories))
width = 0.35
plt.bar(x - width/2, sales_2024, width, label='2024')
plt.bar(x + width/2, sales_2025, width, label='2025')
plt.xticks(x, categories)
plt.xlabel("uarter")
plt.ylabel("Sales")
plt.title("Sales Comparison 2024 vs 2025")
plt.legend()
plt.show()
```

```python
data1 = [10, 12, 14, 16, 18]
data2 = [8, 9, 10, 11, 12]
data3 = [5, 6, 7, 8, 9]
data = [data1, data2, data3]
box = plt.boxplot(data, patch_artist=True, labels=["A", "B", "C"])
colors = ['lightblue', 'lightgreen', 'lightpink']
for patch, color in zip(box['boxes'], colors):
  patch.set_facecolor(color)
plt.title("Colored Box Plots for 3 Groups")
plt.ylabel("Values")
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y1 = [2, 3, 4, 5, 6]
y2 = [1, 2, 1, 2, 1]
plt.stackplot(x, y1, y2, labels=['Dataset 1', 'Dataset 2'], 
colors=['skyblue', 'lightcoral'])
plt.title("Stacked Area Chart")
plt.xlabel("X-axis")
plt.ylabel("Values")
plt.legend(loc='upper left')
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y1 = [1, 2, 3, 4, 5]
y2 = [2, 2, 2, 2, 2]
plt.stackplot(x, y1, y2, labels=['Base', 'Cumulative Layer'], 
colors=['#a1c4fd', '#c2e9fb'])
plt.title("Cumulative Area Chart")
plt.xlabel("X-axis")
plt.ylabel("Values")
plt.legend(loc='upper left')
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y1 = [1, 3, 4, 2, 5]
y2 = [2, 1, 3, 4, 2]
plt.fill_between(x, y1, color='orange', alpha=0.5, label='Series 1')
plt.fill_between(x, y2, color='blue', alpha=0.5, label='Series 2')
plt.title("Overlapping Area Charts with Transparency")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.show()
```
