```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv(r"Z:\ITSkills\sales_data.csv")
months = df['Month'].tolist()
sales = df['Sales'].tolist()
plt.plot(months, sales, marker='o', color='blue')
plt.title('Monthly Sales Data')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.grid(True)
plt.show()
```

```python
df = pd.read_csv(r"Z:\ITSkills\sales_data.csv")
months = df['Month'].tolist()
sales = df['Sales'].tolist()
plt.bar(months, sales, color='green')
plt.title('Monthly Sales Data - Bar Chart')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.tight_layout()
plt.show()
```

```python
df = pd.read_csv(r"Z:\ITSkills\sales_data.csv")
months = df['Month'].tolist()
sales = df['Sales'].tolist()
colors = ['red' if sale == max(sales) else 'blue' for sale in sales]
plt.bar(months, sales, color=colors)
plt.title('Monthly Sales Data - Bar Chart')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.tight_layout()
plt.show()
```

```python
df = pd.read_csv(r"Z:\ITSkills\sales_data.csv")
months = df["Month"].tolist()
sales = df["Sales"].tolist()
plt.pie(sales, labels=months, autopct="%1.1f%%", startangle=140)
plt.title("Sales Contribution by Month")
plt.axis("equal")
plt.tight_layout()
plt.show()
```
